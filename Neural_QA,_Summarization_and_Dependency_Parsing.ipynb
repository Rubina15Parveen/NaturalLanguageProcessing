#Q1: Neural Question Answering

```python
from transformers import pipeline
```

```python
name1 = "deepset/bert-base-cased-squad2"
name2 = "deepset/roberta-base-squad2"
name3 = "deepset/roberta-base-squad2-distilled"
```

```python
model_bert = pipeline("question-answering", model=name1)
model_roberta = pipeline("question-answering", model=name2)
model_distilled = pipeline("question-answering", model=name3)
```

```python
print(model_bert(question="What color is the sky?", context="The sky is blue."))

print("________________________________________________________________________________")

print(model_roberta(question="What color is the sky?", context="The sky is blue."))
print("________________________________________________________________________________")

print(model_distilled(question="What color is the sky?", context="The sky is blue."))
print("________________________________________________________________________________")
```

#Q1.1

```python
question1 = "What two railroads have railyards in the city of Fresno?"
context1 = """Passenger rail service is provided by Amtrak San Joaquins. The main passenger rail station is the recently renovated historic Santa Fe Railroad Depot in Downtown Fresno. The Bakersfield-Stockton mainlines of the Burlington Northern Santa Fe Railway and Union Pacific Railroad railroads cross in Fresno, and both railroads maintain railyards within the city; the San Joaquin Valley Railroad also operates former Southern Pacific branchlines heading west and south out of the city. The city of Fresno is planned to serve the future California High Speed Rail."""
print("BERT")
print(model_bert(question=question1, context=context1))
print("________________________________________________________________________________")

print("RoBERTa")
print(model_roberta(question=question1, context=context1))
print("________________________________________________________________________________")

print("RoBERTa Distilled")
print(model_distilled(question=question1, context=context1))
print("________________________________________________________________________________")

V1 = model_bert(question=question1, context=context1)
V2 = model_roberta(question=question1, context=context1)
V3 = model_distilled(question=question1, context=context1)
print(round(V1['score'],4),V1['answer'])
print(round(V2['score'],4),V2['answer'])
print(round(V3['score'],4),V3['answer'])
```

```python
model_sm = spacy.load("en_core_web_sm")
model_trf = spacy.load("en_core_web_trf")
```

```python
doc = model_sm("This is a sentence.")
for token in doc:
  print(token.text, token.dep_)
```

```python
doc = model_trf("This is a sentence.")
for token in doc:
  print(token.text, token.dep_)
```
